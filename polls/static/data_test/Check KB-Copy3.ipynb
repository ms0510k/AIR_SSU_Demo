```python
import re
import os
import sys
import csv
import pandas as pd
import numpy as np
import datetime
from datetime import timedelta
```

```python

root_dir = "."

user_list = ['park', 'kim', 'choi']

device_list = ['airconditioner', 'aircleaner', 'bulb', 'robotcleaner', 'tv']

season_list = ['spring', 'summer', 'fall', 'winter']

week_list = ['weekdays', 'weekend']

choose_list = []


#* len(device_list) * len(season_list) * len(week_list)) :

for i in range(len(user_list)):
    for j in range(len(device_list)):
        for k in range(len(season_list)):
            for n in range(len(week_list)):
                choose_list.append(
                    root_dir + "/" + user_list[i] + "/" + device_list[j] + "/" 
                    + season_list[k] + "/" + week_list[n])

for i in choose_list:
    print(i)
```

```python
for i in choose_list:
    if not os.path.exists(i):
        os.makedirs(i)
```

```python
file_list = []

def search_dir(dirname, file_list):
    for (path, dir, files) in os.walk(dirname):
        for filename in files:
            ext = os.path.splitext(filename)[-1]
            if ext == '.csv':
                #print("%s/%s" % (path, filename))
                tmp = path + '/' + filename
                #print(tmp)
                file_list.append(tmp)
                
search_dir(choose_list[0], file_list)

print(choose_list[0])
print("list : \n")
for i in file_list:
    print(i)
```

```python
file_list2= [] 
for i in file_list:
    file_list2.append(os.path.abspath(i))
    
for i in file_list2:
    print(i)
```

```python
df_list = []
count = 0
header_name = ["Time", "UID", "Device", "Prob", "Season", "Week", "Weather"]

for i in file_list2:
#     print(i)
    df_list.append(pd.read_csv(i, delimiter='\t', header=None, names=header_name, index_col=None, encoding='CP949', engine="python"))

print(len(df_list))
    
for i in df_list:
    print(i)
```

```python
new = choose_list[0] + "/tmp"
for i in range(10):
    new2 = new+str(i) +".csv"
    print(new2)
    df_list[0].to_csv(new2, sep='\t', index=None, encoding='utf-8')
```
