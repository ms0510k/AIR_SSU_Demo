```python
def format_time(time_tmp):
    s = time_tmp.strftime('%Y_%m_%d_%H_%M_%S_.%f')
    tail = s[-7:]
    f = round(float(tail), 2)
    temp = "%.2f" % f
    return "%s%s" % (s[:-7], temp[2:])
```

```python
def parse_intention_time(start, end):    
    intention_time =[]
    for s, e in zip(start, end):
        s = s.replace(second=0, microsecond=0)
        e = e.replace(second=0, microsecond=0)
        intention_time.append(s)
    return intention_time
```
